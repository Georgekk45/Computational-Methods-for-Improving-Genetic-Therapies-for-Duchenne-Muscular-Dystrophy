```
# Read HBB and HEXA file
df5 = pd.read_excel("dataset.xlsx")
df5 = df5.drop('Spacer sequence', axis=1)
df5.rename(columns={"3' extension":"3extension"}, inplace=True)
df5['Mean of correct edit'] = df5.iloc[:,1:3].mean(axis=1)
```

```
df5 = df5.drop(['% correct edit (w/o indels), rep 1', '% correct edit (w/o indels), rep 2', '% correct edit (w/o indels), rep 3'], axis=1)
```

```
# Split each sequence to columnns
df6 = df5['3extension'].apply(lambda x: [z for y in x for z in y.strip("'").lower().split()])
```

```
mtrx = df6.apply(pd.Series)
```

```
def vector_race(row):
    nr = row.dropna()
    nrl = nr[1:].tolist() # remove index value first item of list
    midl = int(len(nrl)/2) # midle lst
    midm = midl - 5 # left 
    midp = midl + 5 # right
    mid10 = nrl[midm:midp] # mid 10 items 
    strt = nrl[:15]
    end = nrl[-15:]
    final_row = strt + mid10 + end         
    return final_row
```

```
vectors['test'] = mtrx.apply(lambda row: vector_race(row), axis=1)
```

```
vectors_40mtrx = vectors.test.apply(pd.Series)
```
