## **Fillna = 0 method**

```
# Function to convert a DNA sequence string to a numpy array
# Converts to lower case, changes any non 'acgt' characters to 'n'
import numpy as np
import re
def string_to_array(my_string):
    my_string = my_string.lower()
    my_string = re.sub('[^acgt]', 'z', my_string)
    my_array = np.array(list(my_string))
    return my_array

# Create a label encoder with 'acgtn' alphabet
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(np.array(['a','c','g','t','z']))
```

```
# Function to one-hot encode a DNA sequence string
# Non 'acgt' bases (n) are 0000
# Returns a L x 4 numpy array
from sklearn.preprocessing import OneHotEncoder
def one_hot_encoder(my_array):
    integer_encoded = label_encoder.transform(my_array)
    onehot_encoder = OneHotEncoder(sparse=False, dtype=int)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    onehot_encoded = np.delete(onehot_encoded, -1, 1)
    return onehot_encoded
```

```
# Iterate over the 3' extensions and convert each one to vector
vectors = []
encoded = []
for i in range(len(df5["3extension"])):
    test_sequence = df5["3extension"][i]
    encoded = one_hot_encoder(string_to_array(test_sequence))
    vectors.append(encoded)
    print(vectors)
```

```
# Convert arrays to dataframe
arr = np.array(vectors)
vectors_df = pd.DataFrame(data=arr.flatten())
vectors_df.columns = ["Vectors"]
```

```
df5['Vectors'] = vectors_df["Vectors"]
```

```
# Flatten nested lists
df4 = pd.DataFrame([[item for sublist in l for item in sublist] for l in df5['Vectors']])
```

```
# Replace NaN values with 0
df4 = df4.fillna(0)
```

```
vectors_90mtrx = df4.apply(pd.Series)
```

```
vectors_90mtrx['PBS length'] = df5['PBS length']
vectors_90mtrx['RT template length'] = df5['RT template length']
vectors_90mtrx['Mean of correct edit'] = df5['Mean of correct edit']
```
